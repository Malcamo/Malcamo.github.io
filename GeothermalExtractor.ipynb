```python
# prompt: i have excel file it holds energy data, the data is split over multipel tabs. i need to load one of the tabs. just give the code

import pandas as pd

# Replace 'your_excel_file.xlsx' with the actual file name
# Replace 'Sheet1' with the name of the sheet you want to load
df = pd.read_excel('2024 v2 Annual Technology Baseline Workbook Errata 7-19-2024.xlsx',
                   sheet_name='Geothermal', skiprows=263, usecols="K:AO",
                   nrows=6)

df.head(2)
```

```python
df.columns
```

```python
df.columns = ['Technology','Scenario'] + list(df.columns[2:])
```

```python
df
```

```python

df = round(df, 2)

df.head(2)
```

```python
df['Cleaned+Technology'] = df['Technology'].str.replace('Geothermal - ','')
```

```python
df['Technology'] = df['Cleaned+Technology']
```

```python
df
```

```python
df2 = df.drop(['Cleaned+Technology'], axis=1)
```

```python
df3 = df2.set_index(['Scenario','Technology']).stack().reset_index()
```

```python
df3.columns = ['Scenario','Technology','Year','Value']
```

```python
df3.head(2)
```

```python
df3.to_csv('geothermal_LCOE_cleaned.csv', index=False)
# https://tinyurl.com/windbristol
```
