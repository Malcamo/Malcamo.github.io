```python
# prompt: i have excel file it holds energy data, the data is split over multipel tabs. i need to load one of the tabs. just give the code

import pandas as pd

# Replace 'your_excel_file.xlsx' with the actual file name
# Replace 'Sheet1' with the name of the sheet you want to load
df = pd.read_excel('2024 v2 Annual Technology Baseline Workbook Errata 7-19-2024.xlsx',
                   sheet_name='Land-Based Wind', skiprows=332, usecols="K:AO",
                   nrows=30)

df.head(2)
```

```python
df.columns
```

```python
df.columns = ['Class+Technology','Scenario'] + list(df.columns[2:])
```

```python
df.head()
```

```python

df = round(df, 2)

df.head(2)
```

```python
df['Cleaned+Technology'] = df['Class+Technology'].str.replace('Land-Based Wind - ','')
```

```python
mask = df['Cleaned+Technology'] == 'Class 4 - Technology 1'
df = df[mask]
print(df)
```

```python
df['Class'] = df['Cleaned+Technology'].str.split(' - ').str[0].str.replace('Class ','').astype(int)
df['Technology'] = df['Cleaned+Technology'].str.split(' - ').str[1]
```

```python
df2 = df.drop(['Class+Technology', 'Cleaned+Technology'], axis=1)
```

```python
df3 = df2.set_index(['Scenario','Class','Technology']).stack().reset_index()
```

```python
df3.columns = ['Scenario','Class','Technology','Year','Value']
```

```python
df3.head(2)
```

```python
df3.to_csv('onshore_wind_LCOE_cleaned.csv', index=False)
# https://tinyurl.com/windbristol
```
